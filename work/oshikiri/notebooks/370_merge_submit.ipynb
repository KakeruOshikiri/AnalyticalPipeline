```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_oshikiri = pd.read_csv("../submit_oshikiri_best.csv")
df_watanabe = pd.read_csv('../../../../data/submit_watanabe.csv')
df_taniguchi = pd.read_csv('../../../../data/submit_taniguchi.csv')
df_kawai = pd.read_csv('../../../../data/submit_kawai.csv')
```

```python
df_tmp = df_oshikiri.copy()
df_tmp['w'] = df_watanabe.target
df_tmp['k'] = df_kawai.target
df_tmp['t'] = df_taniguchi.target
df_tmp.rename(columns={'target':'o'},inplace=True)
df_tmp
```

```python
df_tmp['all_mean'] = df_tmp[['o', 'k', 'w', 't']].mean(axis=1)
df_tmp['all_median'] = df_tmp[['o', 'k', 'w','t']].median(axis=1)
df_tmp
```

```python
df_tmp['owk_mean'] = df_tmp[['o', 'w', 'k']].mean(axis=1)
df_tmp['owt_mean'] = df_tmp[['o', 'w', 't']].mean(axis=1)
df_tmp['okt_mean'] = df_tmp[['o', 'k', 't']].mean(axis=1)
df_tmp['wkt_mean'] = df_tmp[['w', 'k', 't']].mean(axis=1)

df_tmp['owk_median'] = df_tmp[['o', 'w', 'k']].median(axis=1)
df_tmp['owt_median'] = df_tmp[['o', 'w', 't']].median(axis=1)
df_tmp['okt_median'] = df_tmp[['o', 'k', 't']].median(axis=1)
df_tmp['wkt_median'] = df_tmp[['w', 'k', 't']].median(axis=1)

df_tmp['ow_mean'] = df_tmp[['o', 'w']].mean(axis=1)
df_tmp['ok_mean'] = df_tmp[['o', 'k']].mean(axis=1)
df_tmp['ot_mean'] = df_tmp[['o', 't']].mean(axis=1)
df_tmp['wk_mean'] = df_tmp[['w', 'k']].mean(axis=1)
df_tmp['wt_mean'] = df_tmp[['w', 't']].mean(axis=1)
df_tmp['kt_mean'] = df_tmp[['k', 't']].mean(axis=1)
```
