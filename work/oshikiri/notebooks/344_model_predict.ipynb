## 必要なライブラリのimport

```python
import numpy as np
import pandas as pd
import lightgbm as lgb
from sklearn.metrics import mean_squared_error
import pickle
```

```python
# 予測したいデータの読み込み
path_test_preprp = '../../../../data/processed/test_processed_40.csv'

test = pd.read_csv(path_test_preprp)
```

```python
res_41 = pd.read_csv("../submit_oshikiri_41.csv")
res_42 = pd.read_csv("../submit_oshikiri_42.csv")
```

```python
df_tmp = res_41.merge(res_42,on='card_id')
df_tmp
```

```python
df_tmp.query('target_y<-30')

df_tmp['target_tmp'] = df_tmp.apply(lambda row: np.log2(1e-10) if row['target_y']<-30 else (row['target_x']+row['target_y'])/2,axis=1)
df_tmp
```

```python
import matplotlib.pyplot as plt
#plt.yscale('log')
plt.hist(df_tmp.target_x, bins=1000)
plt.hist(df_tmp.target_tmp, bins=1000)
#plt.hist(predictions[predictions<0], bins=100)
```

## 提出用ファイルの作成

```python
sub_df = pd.DataFrame({"card_id":test["card_id"].values})
sub_df["target"] = df_tmp.target_tmp
sub_df.to_csv("../submit_oshikiri_44_2.csv", index=False)
```
