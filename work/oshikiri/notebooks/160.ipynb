```python
import numpy as np
import matplotlib.pyplot as plt
#import japanize_matplotlib
%matplotlib inline
import pandas as pd
```

```python
x = [0,24,25,26,27,30,40,50,75]
y = [3.61505, 3.61475, 3.61470, 3.61487, 3.61516, 3.61494, 3.61510, 3.61487, 3.61838]
```

```python
df = pd.DataFrame(x,columns=['n'])
df
```

```python
df['private'] = y
df['cv'] = [3.65198, 3.65165 , 3.65178,3.65167 ,3.65232 ,3.65212 ,3.65191 ,3.65172 ,3.65480 ]
df['public'] = [3.69659, 3.69666,3.69736,3.69633,3.69728, 3.69699,3.69667, 3.69654,3.69967]
df
```

```python
plt.title('rmse vs Percentage of features used')
plt.ylabel('rmse')
plt.xlabel('Percentage of features used')

plt.plot(100-df.n,df.public-0.08,marker='.',label='public - 0.08')
plt.plot(100-df.n,df.cv-0.036, marker='.',label='CV - 0.036')
plt.plot(100-df.n,df.private,marker='.',label='private')

plt.legend()
```
